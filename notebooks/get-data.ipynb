Make an HTTP request of all packages (datasets) available on Data.gov (limit is 1,000)

```python
response = requests.get('http://catalog.data.gov/api/3/action/package_list?rows=1000&sort=popularity%20desc')
r = response.json() #returns a dictionary
result = r['result']
info = result['results']

df = pd.DataFrame(info)
df
```

Get only the `name` item from the nested `results` list

```python
response = requests.get('https://catalog.data.gov/api/3/action/package_search?facet.field=[%22name%22]&facet.limit=-1&rows=0')

r = response.json() #returns a dictionary
result = r['result']
info = result['facets']

df = pd.DataFrame(info).reset_index()
df
```
